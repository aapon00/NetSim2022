```python
#packages needed 
import pandas as pd
import scipy.stats as s
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("caida.csv")
df
```

```python
from fitter import Fitter
my_dict1={}
grouped = df[:5000].groupby(['SourceIp','DestinationIP_&_Port'])

for _ , group_df in grouped:
    # if less than 30 packets then skip, can't really fit a distribution
    if len(group_df) < 30:
        continue

    gaps = group_df.Time_Stamp.values.tolist()
    gaps1=[gaps[i+1]-gaps[i] for i in range(len(gaps)-1)]
    df2 = pd.DataFrame({'Time_Stamp': gaps1})
    df3 = df2["Time_Stamp"]
    #Fitting distribution to the time gaps
    f = Fitter(df3, distributions=['gamma', 'lognorm', "beta", "burr", "norm"])
    f.fit()
    dist_best = f.get_best(method='sumsquare_error')
    my_dict1[_] = dist_best
    print (my_dict1)
```

```python
df3
```

```python
#converting the dict into dataframe to do analysis
Time_df = pd.DataFrame.from_dict(my_dict1)
Time_df1=Time_df.T
Time1=Time_df1.notnull()
Time1
```

```python
# normalising the distribution values 
Time1.astype({'beta': int, 'lognorm': int})
#getting all the distribution name in one column
D11=Time1.loc[:, 'beta':'lognorm'].replace(True, pd.Series(Time1.columns, Time1.columns,))
D22 = D11.replace(False,np.nan)
D22['Best_Distribution_Time'] = D22.bfill(axis=1).iloc[:, 0]
D33=D22.drop(['beta','lognorm'], axis=1)
D33
```

```python
D33['new_col'] = range(0, len(D33))
D33.reset_index().set_index('new_col')
DTime=D33.reset_index()
DTime.rename(columns = {'index':'DestinationIP_&_Port'}, inplace = True)
Time_DF=DTime[['Best_Distribution_Time']]
Time_DF
```

```python
from fitter import Fitter
my_dict2={}
grouped = df[:5000].groupby(['SourceIp','DestinationIP_&_Port'])

for i , group_df in grouped:
    # if less than 30 packets then skip, can't really fit a distribution
    if len(group_df) < 30:
        continue

    CRC_list = group_df.CRC.values.tolist()
    CRC_list1=[CRC_list[i+1]-CRC_list[i] for i in range(len(CRC_list)-1)]
    df2 = pd.DataFrame({'CRC': CRC_list1})
    df3 = df2["CRC"]
    f = Fitter(df3, distributions=['gamma', 'lognorm', "beta", "burr", "norm"])
    f.fit()
    dist_best = f.get_best(method='sumsquare_error')
    my_dict2[i] = dist_best
    print (my_dict2)
```

```python
df3 = pd.DataFrame.from_dict(my_dict2)
df4=df3.T
D=df4.notnull()
D
```
